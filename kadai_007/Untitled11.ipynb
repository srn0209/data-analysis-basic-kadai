```python
!pip install japanize-matplotlib

import matplotlib.pyplot as plt
import pandas as pd
import japanize_matplotlib  # 日本語ラベル対応

# サンプルデータ
data = {
    '商品番号': ['A001', 'A001', 'A002', 'A002', 'A003', 'A003', 'A003'],
    'カテゴリー': ['弁当', 'デザート', '弁当', '弁当', 'デザート', '弁当', 'デザート'],
    '価格': [500, 300, 550, 600, 350, 520, 400]
}
df = pd.DataFrame(data)

# カテゴリーの出現頻度をカウントして棒グラフ化
df['カテゴリー'].value_counts().plot(kind='bar')
plt.title("カテゴリー別出現頻度")
plt.xlabel("カテゴリー")
plt.ylabel("件数")

plt.show()

# 平均などの集計（例：平均価格）
平均価格 = df.groupby('商品番号')['価格'].mean()
print("商品番号ごとの平均価格：\n", 平均価格)

# 基本統計量（件数、平均、標準偏差など）
統計量 = df.groupby('商品番号')['価格'].describe()
print("\n商品番号ごとの価格統計量：\n", 統計量)
```
